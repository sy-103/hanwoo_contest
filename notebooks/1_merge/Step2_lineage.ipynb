```python
import pandas

train_df   = pandas.read_csv("../../data/processed/step1_death.csv", low_memory=False)
lineage_df = pandas.read_csv("../../data/raw/hanwoo_lineage.csv", low_memory=False)

print(f"train_df: {train_df.shape}")
print(f"lineage_df: {lineage_df.shape}")

# CATTLE_NO 중복 확인
print(f"\nlineage_df CATTLE_NO 중복: {lineage_df['CATTLE_NO'].duplicated().sum():,}")

before_rows = len(train_df)
before_cols = len(train_df.columns)

train_df = train_df.merge(
    lineage_df,
    on="CATTLE_NO",
    how="left"
)

after_rows = len(train_df)
after_cols = len(train_df.columns)

print(f"\n=== 병합 결과 ===")
print(f"병합 전: {before_rows:,}행 {before_cols}열")
print(f"병합 후: {after_rows:,}행 {after_cols}열")
print(f"\nKPN_NO 결측: {train_df['KPN_NO'].isnull().sum():,}")

# 저장
train_df.to_csv("../data/processed/step2_lineage.csv", index=False, encoding="utf-8-sig")
print(f"\n저장 완료: {train_df.shape}")
```
