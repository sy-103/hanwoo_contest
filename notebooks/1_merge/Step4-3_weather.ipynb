```python
import pandas as pd

weather_all = pd.read_csv("../../data/processed/step_4-2_weather.csv")
weather_all["date"] = pd.to_datetime(weather_all["date"])

# =====================================================
# THI 계산 (국립축산과학원 NRC_1971)
# 입력: 일 최고기온(ta_max) + 일 평균상대습도(rhm_avg)
# =====================================================
weather_all["THI"] = (
        (1.8 * weather_all["ta_max"] + 32)
        - (0.55 - 0.0055 * weather_all["rhm_avg"])
        * (1.8 * weather_all["ta_max"] - 26.8)
)

# =====================================================
# THI 등급 분류 (국립축산과학원 한우 기준)
# =====================================================
def thi_grade(thi):
    if thi < 72:   return "양호"
    elif thi < 78: return "주의"
    elif thi < 89: return "경고"
    elif thi < 98: return "위험"
    else:          return "폐사"

weather_all["THI_grade"] = weather_all["THI"].apply(thi_grade)

print("THI 기술통계:")
print(weather_all["THI"].describe())
print("\nTHI 등급 분포:")
print(weather_all["THI_grade"].value_counts())

# 저장
weather_all.to_csv("../data/processed/step_4-3_weather.csv",
                   index=False, encoding="utf-8-sig")
print(f"\n저장 완료: {weather_all.shape}")
```
