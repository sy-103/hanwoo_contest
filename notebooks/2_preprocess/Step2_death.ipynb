```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/processed/2_preprocess/step1_train.csv", low_memory=False)
df["BIRTH_YMD"]  = pd.to_datetime(df["BIRTH_YMD"], errors="coerce")
df["ABATT_DATE"] = pd.to_datetime(df["ABATT_DATE"], errors="coerce")
df["JUDGE_DATE"] = pd.to_datetime(df["JUDGE_DATE"], errors="coerce")

print(f"로드 완료: {df.shape}")

# death 컬럼 확인
print("\n=== death_count ===")
print(f"결측 수: {df['death_count'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['death_count'].describe())
print(f"\n하위 10개:")
print(df['death_count'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['death_count'].nlargest(10).values)
```

```python
print("death_count 100 이상 농장 확인:")
high_death = df[df['death_count'] >= 100][['death_count', 'FARM_UNIQUE_NO']].drop_duplicates('FARM_UNIQUE_NO').sort_values('death_count', ascending=False)
print(f"해당 농장 수: {len(high_death)}")
print(high_death.head(10))

# 결측 처리 방향 확인
print(f"\n결측 518,737개 = death_df에 없는 농장 소속")
print(f"결측 비율: {df['death_count'].isnull().mean()*100:.2f}%")
```

```python
print(f"death_count 0인 행 수: {(df['death_count'] == 0).sum():,}")
print(f"death_count 최솟값: {df['death_count'].min()}")
```

```python
df.to_csv("../../data/processed/2_preprocess/step2_death.csv",
          index=False, encoding="utf-8-sig")
print(f"저장 완료: {df.shape}")
```
