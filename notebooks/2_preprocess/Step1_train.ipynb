```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/processed/1_merge/step5_merge.csv", low_memory=False)
df["BIRTH_YMD"]  = pd.to_datetime(df["BIRTH_YMD"], errors="coerce")
df["ABATT_DATE"] = pd.to_datetime(df["ABATT_DATE"], errors="coerce")
df["JUDGE_DATE"] = pd.to_datetime(df["JUDGE_DATE"], errors="coerce")

print(f"로드 완료: {df.shape}")
print(f"\ntrain 출처 컬럼:")
print(df[["sido","sigungu","eupmyeondong","stn","ABATT_DATE","JUDGE_DATE",
          "JUDGE_SEX","WEIGHT","BACKFAT","REA","WINDEX","WGRADE","INSFAT",
          "YUKSAK","FATSAK","TISSUE","GROWTH","COST_AMT","AGE","BIRTH_YMD",
          "CATTLE_NO","FARM_UNIQUE_NO","LAST_GRADE"]].dtypes)
```

```python
# sido 기본 현황
print("=== sido ===")
print(f"고유값 수: {df['sido'].nunique()}")
print(f"결측 수: {df['sido'].isnull().sum():,}")
print(f"\n분포:")
print(df['sido'].value_counts())
```

```python
# sigungu, eupmyeondong, stn 한번에 확인
for col in ["sigungu", "eupmyeondong", "stn"]:
    print(f"\n=== {col} ===")
    print(f"고유값 수: {df[col].nunique()}")
    print(f"결측 수: {df[col].isnull().sum():,}")
    print(f"상위 10개:\n{df[col].value_counts().head(10)}")
```

```python
for col in ["ABATT_DATE", "JUDGE_DATE"]:
    print(f"\n=== {col} ===")
    print(f"결측 수: {df[col].isnull().sum():,}")
    print(f"범위: {df[col].min()} ~ {df[col].max()}")
    print(f"연도 분포:\n{df[col].dt.year.value_counts().sort_index()}")
```

```python
print("=== JUDGE_SEX ===")
print(f"고유값 수: {df['JUDGE_SEX'].nunique()}")
print(f"결측 수: {df['JUDGE_SEX'].isnull().sum():,}")
print(f"\n분포:\n{df['JUDGE_SEX'].value_counts()}")
```

```python
print("=== WEIGHT ===")
print(f"결측 수: {df['WEIGHT'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['WEIGHT'].describe())
print(f"\n하위 10개:")
print(df['WEIGHT'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['WEIGHT'].nlargest(10).values)
```

```python
print(df[df['WEIGHT'] <= 10][['WEIGHT', 'AGE', 'JUDGE_SEX', 'LAST_GRADE', 'CATTLE_NO']])
```

```python
print("=== BACKFAT ===")
print(f"결측 수: {df['BACKFAT'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['BACKFAT'].describe())
print(f"\n하위 10개:")
print(df['BACKFAT'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['BACKFAT'].nlargest(10).values)
```

```python
# 상위값 확인
print("BACKFAT 50 이상 분포:")
print(f"행 수: {(df['BACKFAT'] >= 50).sum():,}")
print(df[df['BACKFAT'] >= 50][['BACKFAT', 'AGE', 'JUDGE_SEX', 'LAST_GRADE', 'WEIGHT']].sort_values('BACKFAT', ascending=False).head(20))
```

```python
print("=== REA ===")
print(f"결측 수: {df['REA'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['REA'].describe())
print(f"\n하위 10개:")
print(df['REA'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['REA'].nlargest(10).values)
```

```python
print("REA 10 이하 분포:")
print(f"행 수: {(df['REA'] <= 10).sum():,}")
print(df[df['REA'] <= 10][['REA', 'BACKFAT', 'WEIGHT', 'WINDEX', 'AGE', 'JUDGE_SEX', 'LAST_GRADE']].sort_values('REA'))
```

```python
print("=== WINDEX ===")
print(f"결측 수: {df['WINDEX'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['WINDEX'].describe())
print(f"\n하위 10개:")
print(df['WINDEX'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['WINDEX'].nlargest(10).values)
```

```python
print("=== WGRADE ===")
print(f"결측 수: {df['WGRADE'].isnull().sum():,}")
print(f"\n분포:")
print(df['WGRADE'].value_counts())
```

```python
df['WGRADE'] = df['WGRADE'].replace('-99.0', np.nan)
df['WGRADE'] = df['WGRADE'].replace(-99.0, np.nan)

print(f"처리 후 WGRADE 결측 수: {df['WGRADE'].isnull().sum():,}")
print(f"\n분포:")
print(df['WGRADE'].value_counts())
```

```python
for col in ["INSFAT", "YUKSAK", "FATSAK", "TISSUE", "GROWTH"]:
    print(f"\n=== {col} ===")
    print(f"결측 수: {df[col].isnull().sum():,}")
    print(df[col].describe())
```

```python
print(df[df['GROWTH'].isnull()]['LAST_GRADE'].value_counts())
```

```python
print("=== COST_AMT ===")
print(f"결측 수: {df['COST_AMT'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['COST_AMT'].describe())
print(f"\n하위 10개:")
print(df['COST_AMT'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['COST_AMT'].nlargest(10).values)
```

```python
# 하위값 상세 확인
print("COST_AMT 1000 이하 분포:")
print(f"행 수: {(df['COST_AMT'] <= 1000).sum():,}")
print(df[df['COST_AMT'] <= 1000][['COST_AMT', 'WEIGHT', 'AGE', 'JUDGE_SEX', 'LAST_GRADE']].sort_values('COST_AMT').head(20))
```

```python
print(df[df['COST_AMT'] == 0][['COST_AMT', 'WEIGHT', 'AGE', 'JUDGE_SEX', 'LAST_GRADE', 'WGRADE', 'BACKFAT', 'REA']])
```

```python
df.loc[df['COST_AMT'] == 0, 'COST_AMT'] = np.nan

print(f"처리 후 COST_AMT 결측 수: {df['COST_AMT'].isnull().sum():,}")
print(f"처리 후 최솟값: {df['COST_AMT'].min()}")
```

```python
print("=== AGE ===")
print(f"결측 수: {df['AGE'].isnull().sum():,}")
print(f"\n기술통계:")
print(df['AGE'].describe())
print(f"\n하위 10개:")
print(df['AGE'].nsmallest(10).values)
print(f"\n상위 10개:")
print(df['AGE'].nlargest(10).values)
```

```python
# 하위값 확인
print("AGE 6 이하:")
print(df[df['AGE'] <= 6][['AGE', 'WEIGHT', 'JUDGE_SEX', 'LAST_GRADE']].sort_values('AGE').head(10))

# 상위값 확인
print("\nAGE 200 이상:")
print(f"행 수: {(df['AGE'] >= 200).sum():,}")
print(df[df['AGE'] >= 200][['AGE', 'WEIGHT', 'JUDGE_SEX', 'LAST_GRADE']].sort_values('AGE', ascending=False).head(10))
```
